# Input Holidays

```python
#2-5#6-8#9-12
spring = []#March 18-23
for i in range(6):
  spring.append([3,i+18])
sfinal = []#May 20-31
for i in range(12):
  sfinal.append([5,i+20])

#summer break June-Aug 25
sbreak = []
for i in range(30):
  sbreak.append([6,i+1])
for i in range(31):
  sbreak.append([7,i+1])
for i in range(24):
  sbreak.append([8,i+1])

wfinal = []#Dec 21-30   wbreak:Dec-Jan13
for i in range(10):
  wfinal.append([12,i+21])

holiday=[[1,1],[1,21],[2,18]]+spring+[[5,12]]+sfinal+sbreak+[[9,2],[10,14],[11,11],[11,28],[11,29]]+wfinal
#holiday
```

```python
rf['holiday']=[0]*len(rf)
for i in holiday:
  for j in rf[(rf['Month']==i[0])&(rf['Day']==i[1])].index:
    rf.loc[j,'holiday']=1
```

# Time Series Split and Normalization

```python
#May and June Weekends
TS=[]
trainm=[3,4,5]
rftrain = rf.loc[rf['Month'].isin(trainm)]
testm=[6]
rftest = rf.loc[rf['Month'].isin(testm)]
```

```python
traindata = rftrain.iloc[:,4:]
testdata = rftest.iloc[:,4:]
```

```python
trlabel = traindata[['count']]
trainfeature = traindata.drop(['count','Month','Day','en','out','ds','supply'],axis=1)

telabel = testdata[['count']]
testfeature = testdata.drop(['count','Month','Day','en','out','ds','supply'],axis=1)
```

```python
trf_norm = (trainfeature - trainfeature.min()) / (trainfeature.max() - trainfeature.min())
tef_norm = (testfeature - trainfeature.min()) / (trainfeature.max() - trainfeature.min())
trf_norm.fillna(0, inplace = True)
tef_norm.fillna(0, inplace = True)
```

# MLPRegressor

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV
#best_model = MLPRegressor(hidden_layer_sizes=(10,20), activation='tanh',solver='adam',alpha=1,max_iter=15000)
```

```python
from sklearn.model_selection import GridSearchCV

# Create a parameter grid for hyperparameter tuning
param_grid = {
    'hidden_layer_sizes': [(10,20), (20,20), (30,20)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'lbfgs'],
    'alpha': [0.001, 0.01, 0.1,1]
}

# Create an MLPRegressor instance
model = MLPRegressor()

# Perform grid search to find the best hyperparameters
grid_search = GridSearchCV(model, param_grid, cv=5)
grid_search.fit(trf_norm, trlabel)

# Use the best model from grid search
best_model = grid_search.best_estimator_
print(best_model)
```

```python
best_model.fit(trf_norm, trlabel)
pred_train = best_model.predict(trf_norm)
```

```python
r2 = r2_score(pred_train,trlabel)
print ("Train ERROR = ", r2)
```

```python
from sklearn.metrics import mean_squared_error, r2_score,  mean_absolute_error, mean_absolute_percentage_error
import numpy as np
print(f"平均误差(MAE)：{mean_absolute_error(pred_train,trlabel)}")
print(f"均方误差(MSE)：{mean_squared_error(pred_train,trlabel)}")
print(f"根均方误差(RMSE)：{np.sqrt(mean_squared_error(pred_train,trlabel))}")
print(f"平均绝对百分比误差(MAPE)：{mean_absolute_percentage_error(pred_train, trlabel)}")
print(f"测试集R^2：{r2_score(pred_train,trlabel)}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score,  mean_absolute_error, mean_absolute_percentage_error
import numpy as np

def smape(A, F):
    A=np.array(F)
    F=np.array(A)
    tmp = 2 * np.abs(F - A) / (np.abs(A) + np.abs(F))
    #print((np.abs(A) + np.abs(F)))
    len_ = np.count_nonzero(~np.isnan(tmp))
    if len_ == 0 and np.nansum(tmp) == 0: # Deals with a special case
        return 100
    return 100 / len_ * np.nansum(tmp)

print(f"平均误差(MAE)：{mean_absolute_error(pred_test, telabel)}")
print(f"均方误差(MSE)：{mean_squared_error(pred_test, telabel)}")
print(f"根均方误差(RMSE)：{np.sqrt(mean_squared_error(pred_test, telabel))}")
print(f"平均绝对百分比误差(MAPE)：{mean_absolute_percentage_error(pred_test, telabel)}")
print(f"对称平均绝对百分比误差(SMAPE)：{smape(pred_test, telabel)}")
print(f"测试集R^2：{r2_score(pred_test, telabel)}")
```

# Keras self-defined loss function

```python
import tensorflow as tf
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from keras.models import Sequential
from keras.layers import Dense
dataset = tf.data.Dataset.from_tensor_slices((trf_norm.values, trlabel.values))
#for feat, targ in dataset.take(5):
#  print ('Features: {}, Target: {}'.format(feat, targ))
X = trf_norm.to_numpy()
y =  trlabel.to_numpy()
```

```python
def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(40, activation='relu'),
    tf.keras.layers.Dense(4, activation='relu'),
    tf.keras.layers.Dense(1)
  ])
  model.compile(loss='mse', optimizer='adam',metrics=['mae', 'mse'])
  return model
```

```python
model=get_compiled_model()
model.fit(X,y, epochs=50)
```

```python
pred_train = model.predict(trf_norm.to_numpy())
xx=range(0,len(trlabel))
plt.figure(figsize=(20,6))
plt.scatter(xx,trlabel,color="red",label="Sample Point",linewidth=1)
plt.plot(xx,pred_train,color="orange",label="Fitting Line",linewidth=3)
plt.legend()
plt.show()
```
